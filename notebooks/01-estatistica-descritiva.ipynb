## Calcular estatísticas para cada variável qualitativa

### Tips Dataset

```python
tips_variaveis_qualitativas = tips_dict[tips_dict.tipo == 'qualitativa'].variavel.to_list()
```

```python
for var in tips_variaveis_qualitativas:
    for val in tips[var].unique():
        print(f'{var} - {val}')
        display(tips[tips[var] == val].describe())
        print('*' * 100)
```

### Titanic dataset

```python
titanic_variaveis_qualitativas = titanic_dict[titanic_dict.tipo == 'qualitativa'].variavel.to_list()
```

```python
for var in titanic_variaveis_qualitativas:
    for val in titanic[var].unique():
        print(f'{var} - {val}')
        display(titanic[titanic[var] == val].describe())
        print('*' * 100)
```

## Importações

```python
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
from IPython.display import Markdown
```

## Visualização #1

```python
df = pd.read_csv('https://gist.githubusercontent.com/omadson/770934cb0df9e2d2c3e8db031f1139a5/raw/4d181efc035b327d9d66256915951384517c603e/arrecadacao.csv')
df.head()
```

```python
ax = sns.barplot(data=df, y='arrecadacao', x='titulo')
plt.show()
```

```python
### Utilizando valores simples
df = df.assign(arrecadacao_milhao = df.arrecadacao / 1000000)
# ax = sns.barplot(data=df, y='arrecadacao_milhao', x='titulo')
# plt.show()

### Ordenando as barras
df = df.sort_values(by='arrecadacao_milhao', ascending=False)
# ax = sns.barplot(data=df, y='arrecadacao_milhao', x='titulo')
# plt.show()

### Utilizando linhas verticais
sns.set_style('whitegrid')
# ax = sns.barplot(data=df, y='arrecadacao_milhao', x='titulo')
# plt.show()

### Reduzindo tamanho do gráfico, utilizando a mesma cor para as barras 
### e usando barras horizontais 
plt.figure(figsize=(5,2))
ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='tomato')
# plt.show()


### Adicionando valores nas barras horizontais
ax.bar_label(
    ax.containers[0],
    fmt='%.1f',
    label_type='center',
    color='white',
)
# ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='tomato')
# plt.show()

### Adicioando título e labels nos eixos
plt.xlabel('Arrecadação total (em milhões de dólares)')
plt.ylabel('Filme')
plt.title('Arrecadação de filmes')
# ax = sns.barplot(data=df, x='arrecadacao_milhao', y='titulo', color='tomato')
# plt.show()

### Adicionando uma linha vertical na origem e removendo bordas
ax.set(frame_on=False)
ax.axvline(0, color='k', clip_on=False)

plt.show()
```

```python
df_penguins = sns.load_dataset('penguins')
df_penguins.head()
```

```python
ax = sns.histplot(data=df_penguins, x='body_mass_g')
ax.grid(False)
plt.show()
```

```python
### Utilizando linhas verticais e horizontais
sns.set_style('whitegrid')
ax = sns.histplot(data=df_penguins, x='body_mass_g')

### Melhorando o título e os labels dos eixos
plt.ylabel('Quantidade')
plt.xlabel('Massa (g)')
plt.title('Distribuição da massa dos penguins')

### Adicionando uma linha a origem e removendo as bordas
ax.set(frame_on=False)
ax.axhline(0, color='k', clip_on=False)
plt.show()
```

```python
df_tips = sns.load_dataset('tips')
df_tips.head()
```

```python
ax = sns.boxplot(x='day', y='total_bill', data=df_tips)
ax.grid(False)
```

```python
### Utilizando linhas verticais
sns.set_style('whitegrid')

### Reduzindo o tamanho da figura e utilizando uma mesma cor para as barras
plt.figure(figsize=(6, 3))
ax = sns.boxplot(x='day', y='total_bill', data=df_tips, color='cornflowerblue')

### Melhorando o título e os labels dos eixos
plt.xlabel('Dia da semana')
plt.ylabel('Conta (USD)')
plt.title('Distribuição de gorjetas ao longo da semana')
plt.show()
```
